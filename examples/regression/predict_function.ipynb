```python
import numpy as np
import matplotlib.pyplot as plt

from dlfs.layers import Input, Dense
from dlfs import Sequential
```

## The Problem

In order to get an insight of how our code works, we will solve a simple regression problem.
We have a dataset of 600 points, each with two features $x_0$ and $x_1$, and we want to predict the third
feature:
$y = x_0 \cdot x_1$

### Generating the dataset

```python
f = lambda x_0, x_1: x_0 * x_1
train_x = np.random.uniform(low=-1, high=1, size=(500, 2))
train_y = np.array([f(x_0, x_1) for x_0, x_1 in train_x])
test_x = np.random.uniform(low=-1, high=1, size=(100, 2))
test_y = np.array([f(x_0, x_1) for x_0, x_1 in test_x])

plt.scatter(train_x[:, 0], train_x[:, 1], c=train_y, s=10)
plt.show()
```

### Building the model

The Input layer is the first layer of our model, and it takes the shape of the input data, in this case, two features
($x_0$ and $x_1$). After this, we define the hidden layers, in our example, we have one hidden layer with 8 units. Then,
we define the output layer, which has only one neuron, it will predict the value of $y$.

<img src="./images/neural_network_example_regression.png" width="600" height="400" align="center" />

```python
model = Sequential()
model.add(Input(input_shape=(2,)))
model.add(Dense(8, activation='relu'))
model.add(Dense(1))

model.summary()
```
