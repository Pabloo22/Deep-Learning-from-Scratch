```python
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image

from dlfs.layers import Conv2D
```

```python
img = Image.open('images/car.jpg')
loaded_img = np.array(img) / 255.0
```

```python
eye = np.array([[0, 0, 0],
                [0, 1, 0],
                [0, 0, 0]])

# ed stands for edge detection
ed1 = np.array([[1, 0, -1],
                [0, 0, 0],
                [-1, 0, 1]])

ed2 = np.array([[-1, -1, -1],
                [-1, 8, -1],
                [-1, -1, -1]])

ed3 = np.array([[0, -1, 0],
                [-1, 4, -1],
                [0, -1, 0]])

sharpen = np.array([[-1, 0, -1],
                    [0, 5, 0],
                    [-1, 0, -1]])

gaussian_blur_5x5 = np.array([[1, 4, 6, 4, 1],
                              [4, 16, 24, 16, 4],
                              [6, 24, 36, 24, 6],
                              [4, 16, 24, 16, 4],
                              [1, 4, 6, 4, 1]]) / 256

unsharp_mask5x5 = - np.array([[1, 4, 6, 4, 1],
                            [4, 16, 24, 16, 4],
                            [6, 24, -476, 24, 6],
                            [4, 16, 24, 16, 4],
                            [1, 4, 6, 4, 1]]) / 256
```

```python
def convolve(rgb_img, kernel):
    b = Conv2D.convolve_2d(rgb_img[..., 0], kernel)
    g = Conv2D.convolve_2d(rgb_img[..., 1], kernel)
    r = Conv2D.convolve_2d(rgb_img[..., 2], kernel)
    plt.imshow(np.array([b, g, r]).transpose((1, 2, 0)))
    plt.axis('off')
    plt.show()
```

```python
convolve(loaded_img, eye)
```

```python
convolve(loaded_img, ed1)
```

```python
convolve(loaded_img, ed2)
```

```python
convolve(loaded_img, ed3)
```

```python
convolve(loaded_img, sharpen)
```

```python
convolve(loaded_img, gaussian_blur_5x5)
```

```python
convolve(loaded_img, unsharp_mask5x5)
```
